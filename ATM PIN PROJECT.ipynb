```python
# ATM
savings_account_balance=0
current_account_balance=0
PIN=696969
count=1
while(count==1):
    print("\n---WELCOME TO KAMBOJ ATM!---\n")
    print("->Choose your preferable language:")    
    choice1=input("Enter 1 for English.\nEnter 2 for Hindi.\nEnter 0 to exit:\n")
    choice1=int(choice1)
    if(choice1==0):
        print("Exited!")
        break
    if(choice1==1):
        count_a=1
        while(count_a==1):
            choice2=int(input("\n->Kindly enter your PIN:"))
            if(choice2==PIN):
                print("\n->You want to access SAVINGS or CURRENT account?:")
                choice3=input("Enter 1 for SAVINGS.\nEnter 2 for CURRENT:\n")
                choice3=int(choice3)
                if(choice3==1):            
                    print("\n->You want to WITHDRAW or DEPOSIT:")
                    choice4=input("Enter 1 for WITHDRAW.\nEnter 2 for DEPOSIT:\n")
                    choice4=int(choice4)
                    if(choice4==1):           
                        num1=int(input("Enter the amount you want to withdraw from your SAVINGS account:"))
                        if(savings_account_balance<num1):                    
                            print("Not sufficient balance")
                            break
                        else:               
                            savings_account_balance=savings_account_balance-num1
                            print("₹",num1,"successfully withdrawn from your SAVINGS account and new balance is","₹",savings_account_balance)
                            break
                    elif(choice4==2):                
                        num2=int(input("Enter the amount you want to deposit in your SAVINGS account:"))
                        savings_account_balance=savings_account_balance+num2
                        print("₹",num2,"successfully deposited in your SAVINGS account and new balance is","₹",savings_account_balance)
                        break
                elif(choice3==2):
                    print("\n->You want to WITHDRAW or DEPOSIT:")
                    choice5=input("Enter 1 for WITHDRAW.\nEnter 2 for DEPOSIT:\n")
                    choice5=int(choice5)
                    if(choice5==1):           
                        num3=int(input("Enter the amount you want to withdraw from your CURRENT account:"))
                        if(current_account_balance<num3):                    
                            print("Not sufficient balance")
                            break
                        else:               
                            current_account_balance=current_account_balance-num3
                            print("₹",num3,"successfully withdrawn from your CURRENT account and new balance is","₹",current_account_balance)
                            break
                    elif(choice5==2):                
                        num4=int(input("Enter the amount you want to deposit in your CURRENT account:"))
                        current_account_balance=current_account_balance+num4
                        print("₹",num4,"successfully deposited in your CURRENT account and new balance is","₹",current_account_balance)
                        break
            else:        
                print("Incorrect PIN! Kindly retry!")
    if(choice1==2):
        count_a=1
        while(count_a==1):
            choice2=int(input("\n->कृपया अपना पिन दर्ज करें:"))
            if(choice2==PIN):
                print("\n->आप बचत या चालू खाते तक पहुँचना चाहते हैं?:")
                choice3=input("बचत के लिए 1 दर्ज करें.\nकरंट के लिए 2 दर्ज करें:\n")
                choice3=int(choice3)
                if(choice3==1):            
                    print("\n->आप निकालना या जमा करना चाहते हैं:")
                    choice4=input("निकासी के लिए 1 दर्ज करें.\nजमा के लिए 2 दर्ज करें:\n")
                    choice4=int(choice4)
                    if(choice4==1):           
                        num1=int(input("वह राशि दर्ज करें जो आप अपने बचत खाते से निकालना चाहते हैं:"))
                        if(savings_account_balance<num1):                    
                            print("पर्याप्त संतुलन नहीं")
                            break
                        else:               
                            savings_account_balance=savings_account_balance-num1
                            print("₹",num1,"आपके बचत खाते से सफलतापूर्वक निकासी हो गई है और नया शेष है","₹",savings_account_balance)
                            break
                    elif(choice4==2):                
                        num2=int(input("वह राशि दर्ज करें जो आप अपने बचत खाते में जमा करना चाहते हैं:"))
                        savings_account_balance=savings_account_balance+num2
                        print("₹",num2,"आपके बचत खाते में सफलतापूर्वक जमा हो गया है और नया शेष है","₹",savings_account_balance)
                        break
                elif(choice3==2):
                    print("\n->आप निकालना या जमा करना चाहते हैं:")
                    choice5=input("निकासी के लिए 1 दर्ज करें.\nजमा के लिए 2 दर्ज करें:\n")
                    choice5=int(choice5)
                    if(choice5==1):           
                        num3=int(input("वह राशि दर्ज करें जो आप अपने चालू खाते से निकालना चाहते हैं:"))
                        if(current_account_balance<num3):                    
                            print("पर्याप्त संतुलन नहीं")
                            break
                        else:               
                            current_account_balance=current_account_balance-num3
                            print("₹",num3,"आपके चालू खाते से सफलतापूर्वक निकासी हो गई है और नया शेष है","₹",current_account_balance)
                            break
                    elif(choice5==2):                
                        num4=int(input("वह राशि दर्ज करें जो आप अपने चालू खाते में जमा करना चाहते हैं:"))
                        current_account_balance=current_account_balance+num4
                        print("₹",num4,"आपके चालू खाते में सफलतापूर्वक जमा हो गया है और नया शेष है","₹",current_account_balance)
                        break
            else:        
                print("गलत पिन! कृपया पुनः प्रयास करें!")
```
